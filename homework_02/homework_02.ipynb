## Задание 1: Получение однозначного числа

Пользователь вводит целое число, программа складывает все цифры числа, с полученным числом — то же самое, и так до тех пор, пока не получится однозначное число.

<br>

**Примеры:**
* `545   -> 5`
* `12345 -> 6`

````python
value = ''

while True:
    value = input("Введите целое число: ")

    if value.isdigit():
        break
    else:
        print("Нужно ввести целое число")

while True:
    result = 0

    for i in value:
        result += int(i)

    if result < 10:
        break
    else:
        value = str(result)

print(result)

````

## Задание 2: Кинотеатр

Дан список списков, каждый вложенный список состоит из 1 и 0, количество вложенных списков — количество рядов. Пользователь вводит, сколько билетов ему требуется. Программа должна найти ряд, где можно приобрести нужное количество билетов (места должны быть рядом). Если таких рядов несколько, то ближайший к экрану (ближайшим считается нулевой ряд). Если таких мест нет, то вывести False.

<br>

**Примеры:**
* `[[0,1,1,0], [1,0,0,0], [0,1,0,0]], 2 -> 1`
* `[[0,1,1,0], [1,0,1,0], [1,1,0,1]], 2 -> False`

````python
rows      = [[0,1,1,0], [1,0,0,0], [0,1,0,0]]
tickets   = ""
found_row = None

while True:
    tickets = input("Введите количество билетов: ")

    if tickets.isdigit():
        break
    else:
        print("Нужно ввести целое число")

tickets = int(tickets)

for row_number, seats in enumerate(rows):
    # кол-ов свободных мест подряд
    free_seats_in_row = 0

    if len(seats) < tickets:
        continue

    for seat in seats:
        # занято
        if seat:
           free_seats_in_row = 0
           continue

        free_seats_in_row += 1

        if free_seats_in_row == tickets:
            found_row = row_number
            break

    if found_row is not None:
        break

print(found_row)
````

## Задание 3: Алгоритм RLE

Необходимо написать упрощенную версию алгоритма RLE. 
Алгоритм RLE объединяет подряд идущие символы в коэффициент и символ.

<br>

**Примеры:**
* `asssdddsssddd -> 1a3s3d3s3d`
* `aaabbbbccccc  -> 3a4b5c`
* `abcba         -> 1a1b1c1b1a`

````python
string      = input("Введите строку: ")
char_in_row = ""
count       = 1
result      = ""

for idx, char in enumerate(string):
    if idx == 0:
        char_in_row = char
        continue

    if char == char_in_row:
        count += 1
    else:
        result += str(count) + char_in_row
        count = 1
        char_in_row = char

    if idx + 1 == len(string):
        result += str(count) + char

print(result)
````

`## Задание 4: Шифр Цезаря

Реализуйте программу для шифрования текста с помощью шифра Цезаря.  
Шифр Цезаря — это метод шифрования, при котором каждая буква в тексте заменяется буквой, стоящей на фиксированное число позиций дальше в алфавите.

### Требования
1. Программа должна принимать на вход:
   - Строку текста для шифрования
   - Ключ шифра (число от 0 до 25)

2. Правила шифрования:
   - Шифруются только латинские буквы (a-z, A-Z)
   - Сохраняется регистр букв (заглавные остаются заглавными)
   - Пробелы, цифры и знаки препинания остаются без изменений
   - Сдвиг циклический (после 'z' идет 'a', после 'Z' идет 'A')

### Примеры

* `"Hello World", 3      -> "Khoor Zruog"`
* `"Python is great!", 5 -> "Udymts nx lwjfy!"`
* `"XYZ abc", 3          -> "ABC def"`
* `"Test 123", 1         -> "Uftu 123"`

### Алгоритм
* Для каждого символа в строке:
   - Если это латинская буква, найти её позицию в алфавите
   - Добавить к позиции значение ключа
   - Если результат больше 25, использовать остаток от деления на 26
   - Преобразовать обратно в букву, сохранив регистр
   - Если это не буква, оставить символ без изменений


````python
input_str = input("Введите текст: ")
result    = ""

# a - 97, z - 122
unicode_small_chars = range(97, 123)
# A - 65, Z - 90
unicode_capital_chars = range(65, 91)

key = None

while True:
    key = input("Введите ключ шифра(целое число в диапазано от 1 до 25): ")

    if not key.isdigit():
        print("Нужно ввести целое число")
        continue
    elif 1 > int(key) or int(key) > 25:
        print("Число должно быть в диапазон от 1 до 25")
        continue

    key = int(key)
    break

for char in input_str:
    char_code      = ord(char)
    encrypted_char = char
    chars_range    = None

    if char_code in unicode_capital_chars:
        chars_range = unicode_capital_chars
    elif char_code in unicode_small_chars:
        chars_range = unicode_small_chars
    else:
        result += encrypted_char
        continue

    idx = chars_range.index(char_code)
    replaced_idx = idx + key if idx + key <= len(chars_range) - 1 else idx + key - len(chars_range)
    encrypted_char = chr(chars_range[replaced_idx])

    result += encrypted_char


print(result)
````

## Задание 5: Табель успеваемости

Пользователь в бесконечном цикле (пока не будет введена пустая строка) вводит строки вида: 'название предмета' 'фамилия ученика' 'оценка'. После окончания ввода программа выводит в консоль название предмета, далее список учеников и все их оценки в виде таблицы.

<br>

**Примеры:**

<br>

*Ввод:*
```python
Математика Иванов  5
Математика Иванов  4
Литература Иванов  3
Математика Петров  5
Литература Сидоров 3
Литература Петров  5
Литература Иванов  4
Математика Сидоров 3
Математика Петров  5
```

<br>

*Результат:*
```python
Математика # вводим название предмета

# выводим список учеников и их оценки
Иванов  5 4
Петров  5 5
Сидоров 3
```

```python
Литература # вводим название предмета

# выводим список учеников и их оценки
Иванов  3 4
Сидоров 3
Петров  5
```

````python
print("=============== Начало ввода данных журнала ===================")
subjects = {}
max_length = 0

while True:
    data = input("Введите через пробел 'Название предмета', 'Фамилию ученика' и 'Оценку': ").split(" ")
    data = [item for item in data if item]

    if len(data) == 0:
        break

    if len(data) != 3:
        print("Неверный формат: должно быть три значения('Название предмета', 'Фамилия ученика' и 'Оценка') разделенных одним пробелом")
        continue

    subject, person, score = data

    if int(score) > 5 or int(score) < 1:
        print("Оценка должна быть от 1 до 5")
        continue

    if subject not in subjects:
        subjects[subject] = {}

    if person not in subjects[subject]:
        subjects[subject][person] = []

    subjects[subject][person].append(score)

    if len(person) > max_length:
        max_length = len(person)

for subject in subjects.keys():
    print(f"++++++++ {subject} +++++++++ \n")

    format_string = "{:<" + str(max_length + 3) + "} {:<5}"

    print(format_string.format("Person", "Score"))
    print("-" * 30)

    for person in subjects[subject].keys():
        print(format_string.format(person, " ".join(subjects[subject][person])))

    print("\n\n")






````
