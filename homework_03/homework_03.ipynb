## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
from math import sqrt, floor


def convert_case(string:str)->str:
    result = ''

    if string[0].isupper():
        for idx, char in enumerate(string):
            if char.isupper():
                result += f'_{char.lower()}' if idx != 0 else char.lower()
            else:
                result += char
    else:
        for idx, char in enumerate(string):
            if (idx > 0 and string[idx-1] == "_") or idx == 0:
                result += char.upper()
            elif char == "_":
                continue
            else:
                result += char

    return result

print(convert_case("otus_course"))
print(convert_case("PythonIsTheBest"))
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
# Проверяет корректность даты в формате DD.MM.YYYY
def validate_date(date:str)->bool:
    months_length = {
        1: 31,
        2: 28,
        3: 31,
        4: 30,
        5: 31,
        6: 30,
        7: 31,
        8: 31,
        9: 30,
        10: 31,
        11: 30,
        12: 31,
    }

    date_parts = date.split(".")

    if len(date_parts) != 3:
        return False

    if len(date_parts[0]) != 2 or len(date_parts[1]) != 2 or len(date_parts[2]) != 4:
        return False

    day, month, year = map(int, date_parts)

    is_leap_year = (year % 4 == 0 and year % 100 != 0) or year % 400 == 0

    if not month in months_length:
        return False

    if day < 1 or day > months_length[month]:
        if not(is_leap_year and month == 2 and day == 29):
            return False

    return True

print(validate_date("29.02.2000"))
print(validate_date("29.02.2001"))
print(validate_date("31.04.1962"))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
import math

def is_simple(digit:int)->bool:
    digit = abs(digit)

    if digit == 2:
        return True

    if digit < 2 or digit % 2 == 0:
        return False

    i = 3

    while i <= round(math.sqrt(digit)):
        if digit % i == 0:
            return False
        i += 1

    return True


for i in range(2, 100):
    if is_simple(i):
        print(i)



```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
users = {}

def input_first_name() -> bool|str:
    while True:
        first_name = input("Введите имя:").strip().capitalize()

        if first_name == '':
            return False

        if not first_name.isalpha():
            print("Имя должно содержать только буквы")
            continue

        return first_name

def input_last_name() -> bool|str:
    while True:
        last_name = input("Введите фамилию:").strip().capitalize()

        if last_name == '':
            return False

        if not last_name.isalpha():
            print("Фаимилия должна содержать только буквы")
            continue

        return last_name

def input_age() -> bool|int:
    while True:
        age = input("Введите возраст:").strip()

        if age == '':
            return False

        age = int(age)

        if age < 18 or age > 65:
            print("Возраст должен быть в диапазоне от 18 до 60")
            continue

        return age

def input_id() -> bool|str:
    while True:
        user_id = input("Введите ID:").strip()

        if user_id == '':
            return False

        if not user_id.isdigit() or len(user_id) > 8:
            print("ID должен являться целым числом не более 8 знаков")
            continue

        user_id = user_id.zfill(8)

        if user_id in users:
            print("Такой ID уже есть")
            continue

        return user_id

def print_users():
    format_row = "|{:<8}|{:<15}|{:<15}|{:<7}|"
    divider    = f"+{'-' * 8}+{'-' * 15}+{'-' * 15}+{'-' * 7}+"

    # +-----+------+
    # | asd | asda |
    # +-----+------+

    print(divider)
    print(format_row.format("ID", "Имя", "Фамилия", "Возраст"))
    print(divider)

    for user_id in users.keys():
        print(format_row.format(user_id, *users[user_id]))
        print(divider)

i = 1
while True:
    print(f"================ Начало ввода данных пользователя # {i} ====================")

    first_name = input_first_name()
    if first_name == False:
        print_users()
        break

    last_name = input_last_name()
    if last_name == False:
        print_users()
        break

    age = input_age()
    if age == False:
        print_users()
        break

    user_id = input_id()
    if user_id == False:
        print_users()
        break

    users[user_id] = (first_name, last_name, age)

    print(f"============== Данные пользователя # {i} успешно добавлены ==================")
    i += 1

```
