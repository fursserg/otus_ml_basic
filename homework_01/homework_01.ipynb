## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
valid = False
value = ""

while not valid:
    value = input("Введите пятизначное число:")

    if not value.isdigit():
        print("Ввод должен являться числом")
    elif len(value) != 5:
        print("Число должно быть пятизначным")
    else:
        valid = True
else:
    print(value[0] + value[1:4][::-1] + value[-1])
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
value = input("Сколько дней до отпуска:")

if not value.isdigit():
    print("Похоже, уже пора отдохнуть. Нужно ввести число")


value = int(value)

if value == 0:
    print("Поздравляю, твой отпуск уже начался!")
elif value < 5:
    print("Отпуск начинается до ближайших выходных")
else:
    days = (value // 7)*2

    # подразумевается, что начало отпуска не может попасть на выходные
    if value % 7 >= 5:
        days += 2

    print(f"Выходных до отпуска: {days}")
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
lengthValid, widthValid, wishValid = False, False, False
width, length, wish, square = 0, 0, 0, 0

while not lengthValid:
    length = input("Введите длину шоколадки в плитках: ")

    if length.isdigit() and int(length) > 0:
        length = int(length)
        lengthValid = True
        break
    else:
        print("Введите целое число больше ноля")

while not widthValid:
    width = input("Введите ширину шоколадки в плитках: ")

    if width.isdigit() and int(width) > 0:
        width = int(width)
        widthValid = True
        break
    else:
        print("Введите целое число больше ноля")

square = width * length

while not wishValid:
    wish = input("Сколько плиток отпотчевать желаете: ")

    if not wish.isdigit() or int(wish) <=0:
        print("Введите целое число больше ноля")
    elif int(wish) > square:
        print(f"Много хочешь. В шоколадке всего {square} плиток")
    else:
        wish = int(wish)
        wishValid = True
        break

print(wish % width == 0 or wish % length == 0)
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
value  = ""
result = ""

while True:
    value = input("Введите число:")

    if not value.isdigit() or int(value) <= 0:
        print("Ввод должен являться целым положительным числом")
        continue
    elif int(value) > 3999:
        print("Число больше 3999 нельзя представить в римской нотации")
        continue
    else:
        break


for i, v in enumerate(value[::-1]):
    print(i, v)

    digit = int(v)

    # единицы
    if i == 0:
        if 0 <= digit <= 3:
            result = "I" * digit
        elif 4 <= digit <= 8:
            result = "V" + ("I" * (digit - 5)) if digit >= 5 else "IV"
        else:
            result = "IX"

        continue
    # десятки
    elif i == 1:
        if 0 <= digit <= 3:
            result = ("X" * digit) + result
        elif 4 <= digit <= 8:
            result = ("L" + ("X" * (digit - 5)) if digit >= 5 else "XL") + result
        else:
            result = "XC" + result

        continue

    # сотни
    elif i == 2:
        if 0 <= digit <= 3:
            result = ("С" * digit) + result
        elif 4 <= digit <= 8:
            result = ("D" + ("C" * (digit - 5)) if digit >= 5 else "CD") + result
        else:
            result = "CM" + result

        continue

    # тысячи
    elif i == 3:
        result = ("M" * digit) + result
else:
    print(result)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
def is_positive_float(s):
    # Пустая строка
    if not s:
        return False

    # Проверяем наличие знака в начале
    if s[0] in '+-':
        if s[0] == '-':
            return False
        s = s[1:]
        if not s:
            return False

    has_dot = False
    has_digits = False

    for i, char in enumerate(s):
        if char == '.':
            # Больше одной точки
            if has_dot:
                return False
            has_dot = True

            if i == 0 or i == len(s) - 1:
                continue
        elif '0' <= char <= '9':
            has_digits = True
        else:
            return False

    # Должна быть хотя бы одна цифра
    return has_digits

# Функция для проверки отрицательных чисел
def is_negative_float(s):
    if not s or s[0] != '-':
        return False

    return is_positive_float(s[1:]) or is_positive_float('+' + s[1:])

# Основная программа
if __name__ == "__main__":
    user_input = input("Введите число: ").strip()

    if is_positive_float(user_input):
        print("✓ Это положительное вещественное число")
    else:
        print("✗ Это НЕ положительное вещественное число")

    if is_negative_float(user_input):
        print("✓ Это отрицательное вещественное число")
```
